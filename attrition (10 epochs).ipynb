```python
# Create y_df with the Attrition and Department columns
y_df=attrition_df[['Attrition','Department']]
y_df
```

```python
# Create a list of at least 10 column names to use as X data
X=attrition_df.drop(columns=['Attrition','Department'])


# Create X_df using your selected columns
X_df=pd.DataFrame(X)

# Show the data types for X_df
X_df.dtypes
```

### Lazy_Encoder

```python
def lazy_Encoder(df):
  from sklearn.preprocessing import OneHotEncoder
  ohe=OneHotEncoder(handle_unknown='ignore', sparse_output=False).set_output(transform='pandas')
  for column in df.columns:
    if df[column].dtype=='O':
      if len(df[column].value_counts())==2:
        new_cols=pd.get_dummies(df[column], dtype=int, drop_first=True)
        df[column]=new_cols
      else :
        new_cols=ohe.fit_transform(df[[column]])
        df=pd.concat([df,new_cols], axis=1).drop(columns=[column])
  return df
```

### Back to Original Instruction Cells

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_attrition_train, y_attrition_test, y_department_train, y_department_test\
 =train_test_split(X_df,y_df['Attrition'],y_df['Department'],random_state=42)
```

```python
y_department_train
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
X_train_enc=lazy_Encoder(X_train)
X_test_enc=lazy_Encoder(X_test)
```

```python
# Create a StandardScaler
ss_model=StandardScaler()

# Fit the StandardScaler to the training data
ss_model.fit(X_train_enc)

# Scale the training and testing data
X_train_scaled_df=pd.DataFrame(ss_model.transform(X_train_enc))
X_test_scaled_df=pd.DataFrame(ss_model.transform(X_test_enc))
```

```python
y_department_train
```

```python
# Create a OneHotEncoder for the Department column
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
ohe.fit(np.array(y_department_train).reshape(-1,1))


# Create two new variables by applying the encoder
# to the training and testing data
y_Department_train_enc=ohe.transform(np.array(y_department_train).reshape(-1,1))
y_Department_test_enc=ohe.transform(np.array(y_department_test).reshape(-1,1))
```

```python
# Create a OneHotEncoder for the Attrition column
ohe=OneHotEncoder(sparse_output=False)


# Fit the encoder to the training data
ohe.fit(np.array(y_attrition_train).reshape(-1,1))

# Create two new variables by applying the encoder
# to the training and testing data
y_Attrition_train_enc=ohe.transform(np.array(y_attrition_train).reshape(-1,1))
y_Attrition_test_enc=ohe.transform(np.array(y_attrition_test).reshape(-1,1))
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
input_nodes=len(X_train_scaled_df.columns)

# Create the input layer
input_layer= layers.Input(shape=input_nodes,name='input_features')

# Create at least two shared layers
shared_layer1=layers.Dense(64,activation='relu', name='shared_layer1')(input_layer)
shared_layer2=layers.Dense(32,activation='relu', name='shared_layer2')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
shared_layer3=layers.Dense(64,activation='relu',name='shared_layer3')(shared_layer2)
shared_layer5=layers.Dense(32,activation='relu',name='shared_layer5')(shared_layer3)

# Create the output layer
Department_output = layers.Dense(3,activation='sigmoid',name='Department_output')(shared_layer5)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
shared_layer4=layers.Dense(32,activation='relu',name='shared_layer4')(shared_layer2)

# Create the output layer
Attrition_output = layers.Dense(2,activation='sigmoid',name='Attrition_output')(shared_layer4)
```

```python
# Create the model
model=Model(inputs=input_layer,outputs=[Department_output, Attrition_output])

# Compile the model
model.compile(optimizer='adam',
              loss={'Department_output':'binary_crossentropy',
                    'Attrition_output':'binary_crossentropy'},
              metrics= {'Department_output':'accuracy',
                    'Attrition_output':'accuracy'},
              )

# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(X_train_scaled_df,
          {'Department_output': y_Department_train_enc,
           'Attrition_output': y_Attrition_train_enc},
          epochs=10,
          batch_size=32,
          validation_split=0.2
          )
```

```python
# Evaluate the model with the testing data
test_results=model.evaluate(X_test_enc,
 {'Department_output': y_Department_test_enc, 'Attrition_output': y_Attrition_test_enc})
test_results
```

```python
# Print the accuracy for both department and attrition
print(f"Department Accuracy : {test_results[-2]}")
print(f"Attrition Accuracy : {test_results[-1]}")
```
